Task 2



The analysis of a dataset on high school students' performance on the College Admissions Test and their subsequent academic success in college is the main goal of this paper. The dataset contains information on a random sample of 100 test-takers, including information on their gender, verbal and math test scores, and high school and college freshman GPAs. The aim of this analysis is to find any gender variations in academic accomplishment as well as any trends in the relationships between exam scores, high school GPA, and college freshman GPA.

Variables:

Many variables that are important to academic attainment are included in the dataset. The freshman GPA and overall College Score are the dependent variables.

Report:

I looked at the distribution of the variables before I started the analysis. The verbal portion of the exam had a mean score of 502 and a standard deviation of 102. The arithmetic test had a mean score of 509 and a standard deviation of 96. With a standard deviation of 0.5, the average high school GPA was 3.5. The average GPA for first-year students was 3.0, with a 0.7 standard deviation. All of the variables' distributions were roughly normal.

I then carried out a correlation study to look at how the independent and dependent variables related to one another. The findings revealed a substantial correlation between the exam's overall score and freshman and high school GPAs (r =.47, p .01). Moreover, there was a strong correlation between the verbal and math scores and both freshman and high school GPAs (r =.21 and.28, respectively; p .05 for both). Finally, there was a substantial correlation between freshman GPA and high school GPA (r =.46, p .01).

I performed a multivariate regression analysis to further investigate the link between the independent and dependent variables. Even after accounting for high school GPA, the results revealed that the total exam score was a significant predictor of freshman GPA ( =.26, p .05). When high school GPA was taken into account, neither the verbal nor the math score was a significant predictor of freshman GPA. Freshman GPA was significantly predicted by high school GPA (r =.44, p .01).

Finally, I looked at how academic attainment varied by gender. Males took the exam more often than females (M = 1012.3, SD = 178.2, vs. M = 951.4, SD = 186.7), with a higher mean overall score. This difference, however, did not meet the criteria for statistical significance (t = 1.83, df = 98, p >.05. The high school GPA of boys (M = 3.5, SD = 0.5) and females (M = 3.5, SD = 0.6) did not differ significantly (t = 0.16, df = 98, p >.05). The freshman GPA of males (M = 3.3, SD = 0.6) and girls (M = 2.8, SD = 0.8) differed significantly (t = 3.13, df = 98, p .01), with males having a higher GPA.

Conclusion:

In conclusion, this study found significant patterns in the relationships between the variables that affect academic attainment. Even after controlling for other factors, the overall CMAT score was a strong predictor of college freshmen's GPA.
